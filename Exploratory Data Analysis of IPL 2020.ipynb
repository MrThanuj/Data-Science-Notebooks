### The 2020 Indian Premier League, also known as IPL 13 and branded as Dream11 Indian Premier League 2020, was the thirteenth season of the IPL, a professional Twenty20 cricket (T20) league established by the Board of Control for Cricket in India (BCCI) in 2007.

# <font color='red'> Importing libraries, Loading data, and Cleaning data</font> 

## Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Loading the data                              

##### The <a href="https://www.kaggle.com/rakshit17p/ipl-2020-dataset-dubai"> dataset </a> consist of data about IPL match played in 2020. IPL is a professional Twenty20 cricket league founded by the Board of Control for Cricket in India (BCCI) in 2008. 

##### The league consists of 8 teams representing 8 different Indian cities or states, also has  tremendous popularity all over the world.

```python
data = pd.read_excel ('archive\Ipl_2020.xlsx')
```

## Renaming the columns name

```python
data=data.rename(columns={'Match No':'Match_No','Match in FullForm': 'Match_FullForm','Time India (IST)':'Time_India(IST)','Time UAE':'Time_UAE','Toss Winner':'Toss_Winner','Toss Decision':'Toss_Decision',"Team-1":"Team1","Team-2":"Team2"})
```

## Columns Description

### <font color='red'>The dataset has 19 columns</font>

<b><font color='red'>Match_No</font>  : Number of Matches <br>
<font color='red'>Match</font> : Which team verses which team(Shortform) <br>
<font color='red'>Match_FullForm</font> : Which team verses which team(Fullform) <br>
<font color='red'>Date</font> : Date of match played <br>
<font color='red'>Day</font> : Day of match played <br>
<font color='red'>Time_India (IST)</font> : Match Time in Indian standard time <br>
<font color='red'>Time_UAE</font> : Match Time in United Arab Emirates <br>
<font color='red'>Venue</font> :  Venue(Stadium) where matches played	 <br>
<font color='red'>Team1</font> : One of the current IPL Team 	 <br>
<font color='red'>Team2</font> :	One of the current IPL Team  <br>
<font color='red'>Toss_Winner</font>: Team which won the toss <br>
<font color='red'>Toss_Decision</font> : Toss winner's decision to Bat/Field	 <br>
<font color='red'>dl_applied</font> : (1 or 0)indicates whether the Duckworth-Lewis rule was applied or not. <br>
<font color='red'>winner</font> :  The winner of the match. <br>
<font color='red'>win_by_wickets</font> : Provides the number of wickets by which the team batting second won. <br>
<font color='red'>win_by_runs</font>	: Provides the runs by which the team batting first won <br>
<font color='red'>player_of_the_match</font> : Game Changing play from player gets this award <br>
<font color='red'>Result</font> : The result of the match (‘normal’, ‘tie’) <br>
<font color='red'>Super Over</font> : One Over Eliminator,determine the winner of a limited-overs match in case the scores of both teams at the end of the match are tied<br>

## Heat Map : Null values

```python
plt.figure(figsize=(12,5))
sns.heatmap(data.isnull(),yticklabels=False, annot=True)
plt.xticks(size=15)
plt.show()
```

###### <font color='blue'>Observation: In the above heatmap we can see there is no null values in any attribute of the dataset</font>

## Count plot : Analysing team with the most number of wins in IPL season 13.

```python
plt.figure(figsize=(12,7))
ax = sns.countplot("winner", data = data,order = data['winner'].value_counts().index,palette='viridis')
plt.title("Total number of wins by each team till 2020",size=18)
plt.xticks(rotation=45, ha = 'right',size=13)
plt.yticks(np.arange(13, step=2),size=13)
plt.xlabel('Teams',size=15)
plt.ylabel('Number of matches',size=15)


def annotation_plot(ax,w,h):                                    # function to add data to plot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for p in ax.patches:
        ax.annotate('{0:.1f}'.format(p.get_height()), (p.get_x()+w, p.get_height()+0.2))
annotation_plot(ax,0.08,1)

plt.show()
```

###### <font color='blue'>Observation : We can clearly see that Mumbai Indians(MI) is the team with most number of wins in the season 13, followed by Delhi Capitals(DC) at second position

## Countplot : Winning rate of each team in different Venues.

```python
plt.figure(figsize=(15,8))
ax = sns.countplot("Venue", data = data,hue='winner',palette='rocket',saturation=60.75)
plt.title("Winning rate of each team in different Venues",size=18)
plt.xticks(rotation=45, ha = 'right',size=13)
plt.yticks(size=13)
plt.xlabel('Venue',size=15)
plt.ylabel('Winning rate',size=15)

def annotation_plot1(ax,w,h):                                    # function to add data to plot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for p in ax.patches:
        ax.annotate('{0:.1f}'.format(p.get_height()), (p.get_x(), p.get_height()+0.1))
annotation_plot1(ax,3.28,1)

plt.grid()
plt.show()
```

###### <font color='blue'>Observation : Abu Dhabi has more winning rate, MI has highest and SRH has lowest win rate in Abu Dhabi,In Dubai all the teams has in average same winning rate except RR, RR has lowest win rate in Dubai,coming to Sharjah CSK leads the winning rate and all other teams has same winning rate, but DC has lowest win rate.</font>

## BarPlot : Occurance of Super Over in IPL Venues

```python
plt.figure(figsize=(4,4))
plt.title("Occurance of Super Over in IPL Venues",size=18)
sns.barplot(x='Venue',y='Super Over',hue='Venue',data=data,palette='cool');
plt.yticks(np.arange(0.5, step=0.1),size=12)
plt.xticks(size=10)
plt.ylabel('No of Super Over',size=12)
plt.xlabel('Venue',size=12)
plt.grid()
plt.xticks(rotation=45, ha = 'right')
```

###### <font color='blue'>Observation : Dubai Stadium has witnessed highest number of Super Over and Sharjah being the least.

## BarPlot : Total toss won by Teams

```python
team_wins = data['Toss_Winner'].value_counts()

team_wins_df = pd.DataFrame(columns=["team", "tosswins"])
for items in team_wins.iteritems():
    temp_df1 = pd.DataFrame({
        'team':[items[0]],
        'tosswins':[items[1]]
    })
    team_wins_df = team_wins_df.append(temp_df1, ignore_index=True)
```

```python
team_wins_df
```

```python
plt.figure(figsize=(12,7))
plt.title("Team wise total toss wins",size=20)
ax=sns.barplot(x='team', y='tosswins', data=team_wins_df, palette='inferno',errcolor='6.26',saturation=90.75)
plt.xlabel('Teams',size=15)
plt.ylabel('No of Toss wins',size=15)
plt.yticks(np.arange(12, step=1),size=13)
plt.xticks(size=13)
def annotation_plot(ax,w,h):                                    # function to add data to plot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for p in ax.patches:
        ax.annotate('{0:.0f}'.format(p.get_height()), (p.get_x()+w, p.get_height()+0.2))
annotation_plot(ax,0.19,1)

plt.grid()
plt.show()
```

###### <font color='blue'>Observation : Toss plays vital role in any matches considering dew factor, By above bar graph, it's clear that SRH team has been lucky this season, as they are highest toss winner and KXIP are least times toss winner.

## Players who won 'Player of the Match' and it's count

```python
mvp_ser = data['player_of_the_match'].value_counts()

mvp_ten_df = pd.DataFrame(columns=["player", "wins"])
count = 0
for items in mvp_ser.iteritems():
    temp_df2 = pd.DataFrame({
            'player':[items[0]],
            'wins':[items[1]]
        })
    mvp_ten_df = mvp_ten_df.append(temp_df2, ignore_index=True)
    count += 1    
```

```python
#mvp_ten_df
```

```python
plt.figure(figsize=(12,12))
plt.title("Players who won 'Player of the Match'",size=20)
ax=sns.barplot(x='wins', y='player', data=mvp_ten_df, palette='magma',errcolor='1.26',saturation=20.75);
plt.xticks(np.arange(4, step=1),size=15)
plt.xlabel("No of Man of the match",size=15)
plt.yticks(size=13)
plt.ylabel("Players",size=15)

plt.grid()
plt.show()
```

###### <font color='blue'>Observation : In the above graph x axis presenting number of times the player has won player of match award and y axis presenting player names, in this analysis players named KL Rahul, Ruturaj Gaikwad and AB de villiers has maximum number of player of match award( ie 4 times) and followed by SuryaKumar Yadav, Shikar Dhawan, Ben Stokes, Anrich Nortje,Quinton de kock, Marcus stonis, Yuzvendra Chahal and Sanju Samson has won second highest times(ie 3 times) and rest players has got atmost 1 time.

## Bar Plot : Top 10 Best Performer in 2020 IPL

```python
mvp_ser = data['player_of_the_match'].value_counts()

mvp_ten_df = pd.DataFrame(columns=["player", "wins"])
count = 0
for items in mvp_ser.iteritems():
    if count>9:
        break
    else:
        temp_df2 = pd.DataFrame({
                'player':[items[0]],
                'wins':[items[1]]
            })
        mvp_ten_df = mvp_ten_df.append(temp_df2, ignore_index=True)
        count += 1   
```

```python
plt.figure(figsize=(12,7))
plt.title("Top 10 Best Performer in 2020 IPL",size=20)
sns.barplot(x='wins', y='player', data=mvp_ten_df, palette='magma',errcolor='1.26',saturation=20.75);
plt.xticks(np.arange(4, step=1),size=13)
plt.yticks(size=13)
plt.xlabel("No of Man of the match",size=1)

plt.ylabel("Players",size=15)

plt.grid()
plt.show()
```

###### <font color='blue'>Observation : Top 10 Best player is consider based on the frequently received player of match award. In the above graph x axis presenting number of times the player has won player of match award and y axis presenting player names, in this analysis players named KL Rahul, Ruturaj Gaikwad and AB de villiers has most number of player of match award( ie 4 times) and followed by SuryaKumar Yadav, Shikar Dhawan, Ben Stokes, Anrich Nortje,Quinton de kock, Marcus stonis, Yuzvendra Chahal and Sanju Samson has won second highest times(ie 3 times)

## Boxplot : Each team Won by Runs  & Won by Wickets

### Boxplots are a standardized way of displaying the distribution of data based quartile (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”).

```python
plt.figure(figsize=(7,4))
sns.boxplot(x='winner', y='win_by_runs', data=data,palette='viridis')
plt.title("Each team Won by Runs",size=20)
plt.xticks(rotation=45, ha = 'right',size=15)
plt.xlabel('Teams',size=20)
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(x='winner', y='win_by_wickets', data=data,palette='YlOrBr')
plt.title("Each team Won by wickets",size=20)
plt.xticks(rotation=45, ha = 'right',size=15)
plt.xlabel('Teams',size=20)
plt.show()
```

###### <font color='blue'>Observation : Top most qaurtile called Third qaurtile(Q3)/75th Percentile is representing maximum number of winning runs or wickets, at bottom first quartile(Q1)/25th Percentile is least number of winning runs or wickets and Middle enlongated part is called interquartile range(IQR)/Q2/50th Percentile representing the median of dataset that it is it gives stats of 50 percentile(middle values) of your whole dataset, IQR lie 25th to the 75th percentile.

## Violinplot and Stripplot: Analysis of Win by Runs and Win by Wickets Team wise 


```python
plt.figure(figsize=(15,5))
sns.set(style="darkgrid",palette='deep')
sns.violinplot(x="winner", y='win_by_runs', data=data,inner=None, color=".8")
a=sns.stripplot(x="winner", y="win_by_wickets",hue="win_by_wickets",data=data,size = 9,jitter=0.25)
a.legend(title="win by wickets",loc="center", bbox_to_anchor=(1.06, 0.73), ncol=1)
plt.xticks(rotation=45, ha = 'right',size=15)
plt.yticks(size=15)
plt.xlabel('Teams',size=17)
plt.ylabel('win_by_runs',size=17)
plt.title('Win by Runs and Win by Wickets Team wise',size=20)
plt.show()
```

###### <font color='blue'>Observation : A Violin Plot is used to visualise the distribution of the data and its probability density.Violin plot is similar to Box Plot, and strip plot is scattering data in graph.In the above graph scatter plot is representing win by wickets and violin plot is for representing win by runs.

## Pie Graph : Top Rivalry Matches analysis 

```python
h2h_counts= data.groupby('Match').apply(lambda x:x['Match'].count()).reset_index(name='Match Counts')
top_h2h_order=h2h_counts.sort_values(by='Match Counts',ascending=False)
top_h2h=top_h2h_order[:10]
#print('Head to Head Rivalry\n',top_h2h)
plt.figure(figsize=(7,7))
plt.pie(top_h2h['Match Counts'],labels=top_h2h['Match'],autopct='%1.1f%%', startangle=30)
plt.axis('on')
plt.title('Top 10 most head to head Rivalry Matches Played',size=20)

plt.show()
```

###### <font color='blue'>Observation :MI vs DC, DC vs SRH, RCB vs SRH has high rate of head to head match played each other in this season.

## Scatter Plot : Match Results analysis 

```python
a_list = data['win_by_runs'].tolist()#win_by_runs
a_list=[x + 80 for x in a_list]
plt.figure(figsize=(12,10))
ax = sns.scatterplot(x="Match_No",y='Match',s=a_list,data = data,hue="winner",palette='viridis')
ax.text(60.7,50, "IPL 2020 Final Match(MI vs DC)", horizontalalignment='left', size='medium', color='black', weight='semibold')
ax.text(56.7,52, "Eliminator Match(RCB vs SRH)", horizontalalignment='left', size='medium', color='black', weight='semibold')
ax.text(50,27, "Qualifer 1 Match(DC vs MI)", horizontalalignment='left', size='medium', color='black', weight='semibold')
ax.text(48,11, "Qualifer 2 Match(DC vs SRH)", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.title("Match wise analysis of match results",size=18)
plt.xticks(np.arange(60, step=2),size=15)
plt.yticks(size=10)
plt.xlabel('Match No',size=15)
plt.ylabel('Matchs',size=15)
ax.legend(loc="center", bbox_to_anchor=(1.08, 0.33), ncol=1)
plt.show()
```

###### <font color='blue'>Observation : In the above Scatterplot the data is distributed for analysis of match results with respect to the match number and Team1 vs Team2

## Heat Map: Match results with win count 

```python
match=data.groupby(['Match','winner']).apply(lambda x:x['winner'].count()).reset_index(name='Match Wins')
match_pvt=pd.pivot(match,values='Match Wins',index='Match',columns='winner')
match_pvt.replace(np.NaN,0)
plt.figure(figsize=(7,10))
htmp=sns.heatmap(match_pvt,annot=True,fmt='g',cmap='plasma')
plt.xlabel('Teams',size=17)
plt.ylabel('Matches',size=25)
plt.title('Match results with win count Team wise',size=20)
```

###### <font color='blue'>Observation : We used Heat map here to see the win count of Team when it's played on other teams. As we can see there are many 1's that indicates team 1(X axis) has beaten team 2 once when there was a match team 1 vs team 2(Yaxis) and there are few 2's that indicates team 1 has won twice when team 1 vs team 2 was played.

## HeatMap : Team wise win count at the Venues

```python
venue_win=data.groupby(['Venue','winner']).apply(lambda x:x['winner'].count()).reset_index(name='Match Wins')
venue_win_pvt=pd.pivot(venue_win,values='Match Wins',index='Venue',columns='winner')
venue_win_pvt.replace(np.NaN,0)
plt.figure(figsize=(20,6))
htmp=sns.heatmap(venue_win_pvt,annot=True,fmt='g',cmap='inferno_r')
plt.xticks(size=13)
plt.yticks(size=13)
plt.xlabel('Teams',size=17)
plt.ylabel('Venues',size=17)
plt.title('Team wise win count at the Venues',size=20)

plt.show()
```

###### <font color='blue'>Observation : This Heat map clear insights on how many times the particular team as won in various Venues, in the graph there are three Venues and 8 teams, each team's win count is being analysed in individual Venues.
